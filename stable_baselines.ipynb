```python
import gym
import random
from stable_baselines3 import DQN
```

```python
env = gym.make("CartPole-v1")
```

## before training

```python
episodes = 10
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0
    
    while not done:
        env.render()
        action = random.choice([0, 1])
        n_state, reward, done, info = env.step(action)
        score += reward
    print('Episode:{} Score:{}'.format(episode, score))
```

## training stable_baselines3 dqn (not custom)

```python
env.reset()

model = DQN('MlpPolicy', env, verbose=1)
model.learn(total_timesteps=15000)
```

```python
episodes = 10
for episode in range(1, episodes+1):
    obs = env.reset()
    done = False
    score = 0
    
    while not done:
        action, _states = model.predict(obs, deterministic=True)
        obs, reward, done, info = env.step(action)
        env.render()
        score += reward
    print('Episode:{} Score:{}'.format(episode, score))
```

## training stable_baselines3 dqn (custom)

```python
env = gym.make("CartPole-v0")
env.reset()

model = DQN('MlpPolicy', env, verbose=1, learning_starts=1000,
            target_update_interval=10, train_freq=256, gradient_steps=128, policy_kwargs={'net_arch': [256, 256]})
model.learn(total_timesteps=15000)
```

```python
episodes = 10
for episode in range(1, episodes+1):
    obs = env.reset()
    done = False
    score = 0
    
    while not done:
        action, _states = model.predict(obs, deterministic=True)
        obs, reward, done, info = env.step(action)
        env.render()
        score += reward
    print('Episode:{} Score:{}'.format(episode, score))
```
